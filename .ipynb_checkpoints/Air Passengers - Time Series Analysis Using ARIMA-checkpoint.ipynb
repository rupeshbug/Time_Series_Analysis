### Decomposition of the time series data

Any time series data may consist of the following components:
-  Trend
-  Seasonality
-  Cyclic variations
-  Irregularities

To analyze time series, we need to consider above components and make sure our data is free from all of these components to make a forecast. 

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse


# Multiplicative Decomposition 
multiplicative_decomposition = seasonal_decompose(df['Passengers'], model='multiplicative', period=30)

# Additive Decomposition
additive_decomposition = seasonal_decompose(df['Passengers'], model='additive', period=30)

# Plot
plt.rcParams.update({'figure.figsize': (16,12)})
multiplicative_decomposition.plot().suptitle('Multiplicative Decomposition', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

additive_decomposition.plot().suptitle('Additive Decomposition', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```

```python
plt.figure(figsize = (12,8))
df['Passengers'].plot()
```

From the above graph, we can say that it represents a <b>Multiplicative Model</b> as the data has increasing width and height of the peaks.

### Test for Stationarity

We will use ADF test. This statistical test is done with the following assumptions:  
Null hypothesis: Series is non-stationary  
Alternate hypothesis: Series is stationary

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df['Passengers'])
print(f'ADF Statistic: {result[0]: .3f}')
print(f'p-value: {result[1]:.3f}')
print('Critical Values:')

for key, value in result[4].items():
    print(f'\t{key}: {value:.3f}')
```

p-value > 0.05, so we accept the null hypothesis. The time series data is non-stationary.

### Making a time series stationary using Differencing

```python
# First-order differencing - computing the difference between two consecutive observations in the time series
df_diff = df.diff().dropna()
df_diff.head()
```

```python
# Plot the differenced data

decomp = sm.tsa.seasonal_decompose(df_diff, period=12)
fig = decomp.plot()
plt.show()
```

Trend seems to be removed. However, stationarity is clearly visible.

### ADF test on the differenced data

```python
result = adfuller(df_diff['Passengers'])
print(f'ADF Statistic: {result[0]:.3f}')
print(f'p-value: {result[1]:.3f}')
print('Critical Values:')
for key, value in result[4].items():
    print(f'\t{key}: {value:.3f}')
```

The ADF test on the differenced data shows that the p-value is just above 0.05 and the stationarity test just fails. However, we will still use the time series.

### ARIMA Modeling

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
```

```python
# ACF and PACF plots

fig, ax = plt.subplots(2, 1, figsize=(10,8))
plot_acf(df_diff, lags=20, ax=ax[0])
plot_pacf(df_diff, lags=20, ax=ax[1], method='ywm')
plt.show()
```

#### Selection criteria for the order of ARIMA mode:
-  p: Lag value where the PACF graph cuts off or drops to 0 for the 1st instance
-  d: number of times differencing is carried out to make the time series stationary
-  q: Lag value where the ACF graph crosses the upper confidence interval for the first time (show large drop-off)

The ACF plot show large drop-off at second lag while PACF plot showed significant spikes at the first and second lags. The first-order differencing was used to make the time series stationary. So, we will take the order of the ARIMA model as (2,1,2).
